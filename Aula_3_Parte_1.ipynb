## Teorema de Taylor
$$ f(x+d) = f(x) + d.f'(x) + o(|d|) $$
Se ƒ é diferenciável em torno de x
$$ \lim_{d\to0}{\frac{o(|d|)}{|d|}} $$
Se d é pequeno <b>ƒ(x) + d.ƒ'(x)</b> é uma boa aproximação para <b>ƒ(x+d)</b>.
<br>
<br>
<br>
No lugar de tentar resolver <b>ƒ(x) = 0</b>, escolhemos <b>x0</b> e resolvemos
$$ f(x0) + d.f'(x0) = 0 \to d = \frac{-f(x0)}{f'(x0)} $$
Daí, 
$$ x1 = x0 - \frac{f(x0)}{f'(x0)} $$
Em geral
$$ xk+1 = xk - \frac{f(xk)}{f'(xk)} $$

Quando o Método de Newton funciona, ele é muito mais rápido que o da Bissecção

```julia
"""
    (x, fx, exitflag) = newton(f, a, b)
"""
function newton(f, fd, x;
                   atol = 1e-8, rtol = 1e-8,
                   max_tempo = 10.0, max_iter = 1000,
                   )
    fx = f(x)
    ϵ = atol + rtol * abs(fx)

    iter = 0
    t0 = time()
    Δt = time() - t0
    
    exitflag = :desconhecido
    resolvido = (abs(fx) ≤ ϵ)
    cansado = (iter ≥ max_iter || Δt ≥ max_tempo)
    
    while !(resolvido || cansado)
        fdx = fd(x)
        if abs(fdx) ≤ ϵ
            exitflag = :derivada_nula
            break
        end
        x = x - fx / fdx
        fx = f(x)
        
        iter += 1
        Δt = time() - t0
        resolvido = (abs(fx) ≤ ϵ)
        cansado = (iter ≥ max_iter || Δt ≥ max_tempo)
    end
    
    if resolvido
        exitflag = :sucesso
    else
        if iter ≥ max_iter
            exitflag = :max_iter
        else
            exitflag = :max_tempo
        end
    end
    return x, fx, exitflag #plot(f, a, b)
end
```

```julia
# Um problema que funciona
# Tendo precisão suficiente pra que ele não pare e tempo suficiente pra ele rodar, sempre funciona
f(x) = x^2 - 2
fd(x) = 2x
newton(f, fd, 1.0)
```

```julia
setprecision(512)
f(x) = x^2 - 2
fd(x) = 2x
newton(f, fd, big"1", atol=1e-100, rtol=0.0)
```

```julia
f(x) = x * exp(x) - 1
fd(x) = exp(x) + x * exp(x)
newton(f, fd, 1.0)
```

```julia
n = 48
V = 40_000
p = 1200
function eq_price(i)
    if i == 0
        return p - V / n
    else
        return p - V * (1 + i)^n * i / ((1 + i)^n - 1)
    end
end
```

```julia
Pkg.add("ForwardDiff")
```

```julia
# Não temos a derivada da eq_price(i), para isso, usaremos o pacote "ForwardDiff" que calcula rapidamente a derivada
using ForwardDiff

eq_price_der(i) = ForwardDiff.derivative(eq_price, i)
newton(eq_price, eq_price_der, 1.0)
```

## Fractal de x³ - x

```julia
using Plots

f(x) = x^3 - x
fd(x) = 3x^2 - 1

anim = Animation()
for x = -1.0:0.01:1.0
    plot(f, -1.4, 1.4, leg=false)
    scatter!([x], [f(x)], c=:red, ms=4)
    local x
    for k = 1:3
        xp = x - f(x) / fd(x)
        plot!([x, xp, xp], [f(x), 0, f(xp)], c=:pink)
        scatter!([xp], [f(xp)], x=:pink, ms=2)
        x = xp
    end

    xlims!(-1.4, 1.4)
    ylims!(-1.5, 1.5)
    frame(anim)
end
gif(anim, "exemplo.gif", fps=10)
# Esse gif mostra bem os intervalos em que a convergência vai pra -1, 0 e 1
```

```julia
x = sqrt(3) / 3 - 0.115
x = sqrt(5) / 5 - 1e-4 # Situação que o Método de Newton cai num loop e fica dando voltas sem convergir
plot(f, -1.4, 1.4, leg=false)
scatter!([x], [f(x)], c=:red, ms=4)
local x
for k = 1:10
    xp = x - f(x) / fd(x)
    plot!([x, xp, xp], [f(x), 0, f(xp)], c=:pink)
    scatter!([xp], [f(xp)], x=:pink, ms=2)
    x = xp
end

xlims!(-1.4, 1.4)
ylims!(-1.5, 1.5)
```

```julia
f(x) = x^3 - x
fd(x) = 3x^2 - 1

anim = Animation()
X = []
C = []
colors = [:red, :green, :blue]
r0 = -1.4
rf = -r0
I = r0:0.01:rf
for x = I
    plot(f, r0, rf, leg=false)
    scatter!([x], [f(x)], c=:red, ms=4)
    push!(X, x)
    local x
    for k = 1:10
        xp = x - f(x) / fd(x)
        plot!([x, xp, xp], [f(x), 0, f(xp)], c=:pink)
        scatter!([xp], [f(xp)], x=:pink, ms=2)
        x = xp
    end
    j = argmin(abs.(x .- [-1, 0, 1]))
    push!(C, colors[j])
    for (xi,ci) in zip(X, C)
        plot!([xi,xi], [-1.5, 1.5], c=ci)
    end

    xlims!(r0, rf)
    ylims!(-1.5, 1.5)
    frame(anim)
end
gif(anim, "exemplo.gif", fps=10)
# A partir desse gif, fica ainda mais claro, quando a convergência é para -1, 0 ou 1, basta ver a cor das linhas
# Vermelho -> -1
# Verde -> 0
# Azul -> 1
```

```julia
f(x) = x^3 - x
fd(x) = 3x^2 - 1

anim = Animation()
X = []
C = []
colors = [:red, :green, :blue]
r0 = sqrt(5) / 5
rf = r0 + 0.001
I = r0 .+ exp.(range(log(1e-6),
                     log(rf - r0 - 1e-4), length=100))
for x = I
    plot(f, r0, rf, leg=false)
    scatter!([x], [f(x)], c=:red, ms=4)
    push!(X, x)
    local x
    for k = 1:10
        xp = x - f(x) / fd(x)
        plot!([x, xp, xp], [f(x), 0, f(xp)], c=:pink)
        scatter!([xp], [f(xp)], x=:pink, ms=2)
        x = xp
    end
    j = argmin(abs.(x .- [-1, 0, 1]))
    push!(C, colors[j])
    for (xi,ci) in zip(X, C)
        plot!([xi,xi], [-1.5, 1.5], c=ci)
    end

    xlims!(r0, rf)
    ylims!(-1.5, 1.5)
    frame(anim)
end
gif(anim, "exemplo.gif", fps=10) # Fractal de sqrt(3) / 3
```
