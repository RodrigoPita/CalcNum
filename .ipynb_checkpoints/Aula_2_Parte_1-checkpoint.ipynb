```julia
# O Professor Abel menciona ter feito a função fatorial mais cedo no vídeo, mas como a atividade era só a partir de 1:07h,
# fiz a minha versão dessa função
function fatorial(x)
    if x == 1
        return x
    else 
        return x*fatorial(x-1)
    end
end
```

```julia
function exp1(x, n)
    # 1 + x + x^2/2 + x^3/3! + x^4/4! + ...
    s = 1.0
    for k = 1:n
        s += x^k/fatorial(k)
    end
    return s
end
```

```julia
exp1(5.0, 30) - exp(5.0)
# ficou ruim pois deu fatorial acima de 21 dá overflow, não teve avisos por ter ocorrido na função que eu implementei;
# usando a função "factorial" do próprio Julia, teria aparecido uma mensagem de erro
# não queremos também usar BigInt, pra não gastar mais memória do que o necessário
```

```julia
function exp3(x, n)
    # 1 + x + x^2/2 + x^3/3! + x^4/4! + ...
    s = 1.0
    a = 1.0
    for k = 1:n
        a = a*x/k
        s += a
    end
    return s
end
```

```julia
exp3(5.0, 30)
# não explode por não ter fatorial
```

```julia
exp3(5.0, 32) - exp(5)
```

```julia
# Tentando chegar na funcão exponencial do julia, que só recebe um parâmetro de entrada
# Para isso, faz-se necessário colocar um condição "mágica"
function exp4(x)
    # 1 + x + x^2/2 + x^3/3! + x^4/4! + ...
    s = 1.0
    a = 1.0
    k = 1
    while s + a != s
        a = a*x/k
        s += a
        k += 1
    end
    return s
end
```

```julia
using Plots
X = range(0.0, 10.0, length=100)
Erro = [max(abs(exp4(x) - exp(x))/exp(x), 1e-16) for x in X]
plot(X, Erro, yaxis=:log)
xlabel!("x")
ylabel!("|exp4(x) - exp(x)|/exp(x)")
title!("Erro relativo")
# Esse erro mostra que os algorítmos não são iguais, mas muito próximos
```

```julia
using Plots
X = range(-10.0, 10.0, length=100)
Erro = [max(abs(exp4(x) - exp(x))/exp(x), 1e-16) for x in X]
plot(X, Erro, yaxis=:log)
xlabel!("x")
ylabel!("|exp4(x) - exp(x)|/exp(x)")
title!("Erro relativo")
```

```julia
function exp5(x)
    if x >= 0
        return exp4(x)
    else
        return 1.0/exp4(-x)   
    end
end
```

```julia
using Plots
X = range(-10.0, 10.0, length=100)
Erro = [max(abs(exp5(x) - exp(x))/exp(x), 1e-16) for x in X]
plot(X, Erro, yaxis=:log)
xlabel!("x")
ylabel!("|exp5(x) - exp(x)|/exp(x)")
title!("Erro relativo")
```

$$ \arctan 1 = \frac{\pi}{4} $$

```julia
function pi_atan(n)
    # π ≈ 4 * (1 -1/3 + 1/5 - 1/7 + 1/9 ... (n-ésimo?))
    # n =      1,  2,    3,    4,    5, 
    aux2 = 0 # usado para ajeitar o denominador de cada componente da soma
    pi = 0 # o que queremos retornar
    for k = 1:n
        if k % 2 != 1
            aux1 = -1 # usado para alternar o sinal da soma
        else
            aux1 = 1
        end
        # println("1/$(aux2+k)") ## para verificar se as frações estavam corretas
        pi += aux1 * 4(1/(aux2+k))
        aux2 += 1
    end
    return pi
end
```

```julia
for k = 1:10000000
    println(pi_atan(k))
end
# Com esse número de iterações, consegui aproximar pi com apenas 4 casas decimais
# Não pretendo tentar mais, pois assim já demorou bastante para rodar o código
```
