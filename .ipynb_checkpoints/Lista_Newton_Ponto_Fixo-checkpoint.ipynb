```julia
"""
    (x, fx, exitflag) = newton(f, a, b)
"""
function newton(f, fd, x;
                   atol = 1e-8, rtol = 1e-8,
                   max_tempo = 1000.0, max_iter = 1000000,
                   )
    fx = f(x)
    ϵ = atol + rtol * abs(fx)

    iter = 0
    t0 = time()
    Δt = time() - t0
    
    exitflag = :desconhecido
    resolvido = (abs(fx) ≤ ϵ)
    cansado = (iter ≥ max_iter || Δt ≥ max_tempo)
    
    while !(resolvido || cansado)
        fdx = fd(x)
        if abs(fdx) ≤ ϵ
            exitflag = :derivada_nula
            break
        end
        x = x - fx / fdx
        fx = f(x)
        
        iter += 1
        Δt = time() - t0
        resolvido = (abs(fx) ≤ ϵ)
        cansado = (iter ≥ max_iter || Δt ≥ max_tempo)
    end
    
    if resolvido
        exitflag = :sucesso
    else
        if iter ≥ max_iter
            exitflag = :max_iter
        else
            exitflag = :max_tempo
        end
    end
    return x, fx, exitflag 
end
```

<b>3)</b> Explique como aproximar $ \sqrt[3]{43} $ com o método de Newton e determine um chute inicial problemático.

=======================================================================================================================

Para aproximar $\sqrt[3]{43}$, primeiro você precisa gerar a função ƒ(x) que tem como zero o esse valor, após isso, encontramos a derivada dessa função. Feito essas duas etapas, basta jogar no algoritmo e chutar um $x_{0}$.

```julia
f(x) = x^3 - 43
fd(x) = 3x^2
r1 = newton(f, fd, 15)
r2 = newton(f, fd, 1000)
r3 = newton(f, fd, 1500)
r1[1], r2[1], r3[1]
```

Chutando 1500, por algum motivo ele para antes de chegar no lugar certo, então entendo $x_{0} = 1500$ como um chute inicial problemático.
